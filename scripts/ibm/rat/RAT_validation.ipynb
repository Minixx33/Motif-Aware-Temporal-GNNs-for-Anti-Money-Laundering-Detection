This notebook is for debugging and validating that RAT injection was done correctly on the datasets. Use this script to create the plots to visualize how the dataset is structured.

```python
import pandas as pd

pd.DataFrame({
    "Metric": ["Original Rows", "Injected Rows", "Total"],
    "Count": [
        df['injected_motif'].isna().sum(),
        df['injected_motif'].notna().sum(),
        len(df)
    ]
})
```

```python
import matplotlib.pyplot as plt

sizes = [df['injected_motif'].notna().sum(), df['injected_motif'].isna().sum()]
labels = ["Injected Motifs", "Original Transactions"]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct="%1.4f%%", startangle=140)
plt.title("Proportion of Injected Motifs in Dataset")
plt.show()
```

```python
import matplotlib.pyplot as plt

counts = {
    "Original": df['injected_motif'].isna().sum(),
    "Injected": df['injected_motif'].notna().sum()
}

plt.figure(figsize=(6,4))
plt.bar(counts.keys(), counts.values())
plt.yscale('log')
plt.title("Injected Motifs vs Original Transactions (log scale)")
plt.ylabel("Count (log)")
plt.show()
```
